```python
conjuntos= { "maria", "joana", "pedro", "ana","carlos" }
print(conjuntos)
# agregar nombre

conjuntos.add("jimmy")
print(conjuntos)

conjuntos_a= {"maria", "joana", "pedro", "ana","carlos" }
conjuntos_b= {"fernando", "joana", "martina", "juana","paola" }

# se utilizan operaciones matematicas para uso de coincidencias, repetidos, comparar datos

print("interseccion:", conjuntos_a.intersection(conjuntos_b))
print("union:", conjuntos_a.union(conjuntos_b))
print("diferencia:", conjuntos_a.difference(conjuntos_b))
```
