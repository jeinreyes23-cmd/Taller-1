```python
columnas = [
    "Glucose",
    "BloodPressure",
    "SkinThickness",
    "Insulin",
    "BMI"
]

for columna in columnas:
    print(columna)
    print((df[columna] == 0).sum())
```

```python
df[columnas] = df[columnas].replace(0, np.nan)
```

```python
for columna in columnas:
    df[columna] = df[columna].fillna(df[columna].median())
```
